### Data Cleaning

```python
EDA_df.columns.values
```

```python
#Many columns can be neglected for the calculations 
EDA_df.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country','provstate':'State', 'city':'City', 'attacktype1_txt': 'Attacktype', 'targtype1_txt':'Targettype', 'targsubtype1_txt':'Targetsubtype', 'target1':'Target', 'weaptype1_txt':'Weapon', 'nkill':'kill', 'nwound':'Wound', 'gname':'Group', 'motive':'Motive', 'region_txt':'Region'}, inplace=True)
```

```python
#Unnecessary columns Removal
df = EDA_df[['Year', 'Month', 'Day', 'Country','State', 'City', 'Attacktype', 'Targettype', 'Targetsubtype', 'Target', 'Weapon', 'kill', 'Wound', 'Group', 'Motive', 'Region']]
```

### Terrorist Activities Every Year

```python
x_year = df['Year'].unique()
y_count_years = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 50)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks Each Year')
plt.title('Attack In Years')
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot('Year',data=df,palette='RdYlGn_r',edgecolor=sns.color_palette("YlOrBr", 10))
plt.xticks(rotation=50)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

### Terrorist Activities By Region In Each Year

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year')
plt.ylabel('Number of Attacks')
plt.show()
```
